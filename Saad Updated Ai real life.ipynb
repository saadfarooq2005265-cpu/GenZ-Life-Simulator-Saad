```python
import random

# -------------------------- Profile setup --------------------------
def setup_profile():
    """Get name and age from user safely"""
    name = input("What's your name? ")
    while True:
        age_str = input("How old are you? ")
        try:
            age = int(age_str)
            break
        except:
            print("Please enter a number for your age (e.g., 20).")
    user = {"name": name.strip(), "age": age}
    print(f"\nHey {user['name']} ğŸ‘‹ Welcome to GenZ Life Simulator!\n")
    return user

# -------------------------- Keyword-based chatbot responses --------------------------
responses = {
    "hi": "Hey bestie! ğŸ˜",
    "hello": "Hey bestie! ğŸ˜",
    "how are you": "Iâ€™m vibing, wbu? ğŸ˜Œ",
    "what's up": "Just chilling and ready to play ğŸ˜œ",
    "bye": "Aww, see you soon ğŸ’–",
    "thanks": "Anytime! â¤ï¸"
}

def get_response(user_input):
    """Normalize common phrases and respond"""
    key = user_input.strip().lower()
    if key in responses:
        return responses[key]
    if "hi" in key or "hello" in key:
        return responses["hi"]
    if "how are" in key:
        return responses["how are you"]
    return "Hmmâ€¦ idk what to say ğŸ¤”"

def greet_user(user):
    print(f"âœ¨ Yo {user['name']}! I'm your AI bestie. Type 'play' to start the GenZ Life Simulator or 'bye' to quit")

def end_chat(user):
    print(f"Bye {user['name']}! Catch you later ğŸ’«")

# -------------------------- Game stats and functions --------------------------
stats = {"energy": 100, "points": 0, "mood": "ğŸ™‚"}

def show_status():
    """Clamp energy and display current stats"""
    stats['energy'] = max(0, min(100, stats['energy']))
    print("\n--- Current Status ---")
    print(f"Energy: {stats['energy']} âš¡ | Points: {stats['points']} ğŸ’ | Mood: {stats['mood']}")
    print("----------------------\n")

def update_stats(activity):
    if activity == "study":
        stats["energy"] -= 20
        stats["points"] += 10
        stats["mood"] = "ğŸ˜ (studious)"
        print("You studied hard! +10 points (-20 energy).")
    elif activity == "eat":
        stats["energy"] += 15
        stats["mood"] = "ğŸ˜‹ (satisfied)"
        print("Yummy! +15 energy.")
    elif activity == "play":
        stats["energy"] -= 10
        stats["points"] += 5
        stats["mood"] = "ğŸ˜„ (excited)"
        print("Game on! +5 points (-10 energy).")
    elif activity == "sleep":
        stats["energy"] = 100
        stats["mood"] = "ğŸ˜´ (rested)"
        print("You slept well â€” energy restored to 100.")
    else:
        print("Thatâ€™s not an activity I know ğŸ˜… Try: study / eat / play / sleep")
    stats['energy'] = max(0, min(100, stats['energy']))

def end_game(user):
    print("\nâœ¨ Game Over âœ¨")
    print(f"{user['name']}, you finished with {stats['points']} points and {stats['energy']} energy.")
    print("Summary of mood:", stats['mood'])
    print("Thanks for playing â€” see you next time! ğŸ’–")

# -------------------------- Unique feature: Daily Quote Generator --------------------------
quotes = [
    "Keep slaying, bestie ğŸ’…",
    "Donâ€™t let bad vibes ruin your glow âœ¨",
    "Eat, sleep, vibe, repeat ğŸ˜",
    "One more task, then full chill mode ğŸ§˜",
    "Small wins = big vibes ğŸ’¯"
]

def daily_quote():
    print("\nâ˜€ï¸ Today's Vibe Quote:", random.choice(quotes), "\n")

# -------------------------- Game loop --------------------------
def start_game(user):
    print("\nğŸ® Welcome to GenZ Life Simulator ğŸ®")
    daily_quote()  # unique feature triggered once at start
    while True:
        show_status()
        activity = input("Choose an activity (study / eat / play / sleep) or type 'quit' to end game: ").strip().lower()
        if activity == "quit":
            end_game(user)
            break
        update_stats(activity)
        if stats['energy'] <= 0:
            print("Your energy is 0 â€” you should sleep or eat to continue.")

# -------------------------- Conversation loop (chatbot) --------------------------
def chat_loop(user):
    greet_user(user)
    while True:
        user_input = input("You: ").strip()
        if user_input == "":
            print("Speak to me, bestie ğŸ˜„")
            continue
        lower = user_input.lower()
        if lower == "bye":
            end_chat(user)
            break
        if "play" in lower or "game" in lower:
            choice = input("Ooo want to play GenZ Life Simulator with me? (yes/no): ").strip().lower()
            if choice in ("yes", "y"):
                start_game(user)
                print("\nBack to chat mode. Say 'bye' to exit.")
            else:
                print("Okayy, let's just keep chatting ğŸ˜‡")
        else:
            reply = get_response(user_input)
            print("Bestie:", reply)

# -------------------------- Run everything --------------------------
if __name__ == "__main__":
    print("Starting GenZ Life Simulator...\n")
    user = setup_profile()
    chat_loop(user)
```
